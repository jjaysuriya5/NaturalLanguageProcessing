# **Creating Intents and entities**

```
domain ="""
intents:
- greet
- goodbye
- feedback_rate
- feedback_info
entities:
- email
- information
- location

templates:
  utter_apt_options:
  - buttons:
    - payload: /OnlineAppointment
      title: Online Appointment
    - payload: /FeedBack
      title: FeedBack
    text: How May I Help You
  utter_cheer_up:
  - image: https://i.imgur.com/nGF1K8f.jpg
    text: 'Here is something to cheer you up:'
  utter_city_Which:
  - buttons:
    - payload: /Delhi
      title: Delhi
    - payload: /Lucknow
      title: Lucknow
    - payload: /Kolkata
      title: Kolkata
    - payload: /Agra
      title: Agra
    text: Which area are you looking for the Hospital in?
actions:
- utter_apt_options
- utter_cheer_up
- utter_city_Which
"""
%store domain > '/content/domain.yml'
```

```
nlu = """
## intent:information
- [md sarfaraz](name)
- [1234567890](number)
- [22/09/1992](dob)
- [abc@xyz.com](email)
## intent:greet
- hey
- hello
"""
%store nlu > '/content/data/nlu.md'
```

# **Preparing config file**

```
config = """
# Configuration for Rasa NLU.
# https://rasa.com/docs/rasa/nlu/components/
language: en
pipeline: supervised_embeddings

# Configuration for Rasa Core.
# https://rasa.com/docs/rasa/core/policies/
policies:
  - name: MemoizationPolicy
  - name: KerasPolicy
  - name: MappingPolicy
"""

%store config > '/content/config.yml'
```
